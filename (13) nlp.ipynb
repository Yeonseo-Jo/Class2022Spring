### Tokenization

* string을 단어 수준에서 끊는 것


* tokenize 방법들
1. 방법 (1) = split 메서드 : 띄워쓰기 기반 => 단점 : functuation 구분 못함
2. 방법 (2) : nltk.tokenize 패키지에서 word_tokenize 함수 사용 (->'punkt'로 punctuation들도 구분 가능)
3. 방법 (3) : nltk.tokenize 패키지의 RegexpTokenizer 함수 사용 (=> regular expression을 이용하여 특정 부분 제거 가능) / ex. punctuation들 제거


```python
import nltk
```

```python
# copy a file from github       #첫번째 방식 - Github 사이트에 있는 사이트 url로 파일을 현재 server에 copy한 후, server에 있는 파일을 읽어오는 방식
import os
url = "https://raw.githubusercontent.com/hsnam95/class2022Spring/main/crime_punishment.txt"
os.system("curl " + url + " > crime_punishment.txt")

# read a text file in the server
file = open("crime_punishment.txt")
text = file.read().replace("\n", " ")                          # text라는 변수에 url에 있는 text를 string의 형태로 가져옴
file.close()
```

```python
# or copy/pase text here                # 두번째 방식 - 직접 string의 값을 입력하는 방식
text = 'Here’s to the crazy ones, the misfits, the rebels, the troublemakers, the round pegs in the square holes. \
The ones who see things differently — they’re not fond of rules. \
You can quote them, disagree with them, glorify or vilify them, \
but the only thing you can’t do is ignore them because they change things. \
They push the human race forward, and while some may see them as the crazy ones, we see genius, \
because the ones who are crazy enough to think that they can change the world, are the ones who do.'
```

```python
# write out a text file             # 참고 - string을 text 파일로 만드는 방법 (file open해서 write함수로 내용 써줌 *close 해주기 !!)
file = open("tmp.txt", "w")
file.write(text)
file.close()
```

```python
text.split()       # (1) tokenize 방법 1 = split 메서드 : 띄워쓰기를 기준으로 string을 쪼개줌.
                   # 단점: punctuation과 같은 것들을 구분하지 못함
```

```python
' '.join(text.split())  #split 한것을 ' '로 다시 결합해서 원 text로 만들어줌
```

```python
from nltk.tokenize import word_tokenize       #2. 방법 (2) : nltk.tokenize 패키지에서 word_tokenize 함수 사용 (->'punkt'로 punctuation들 구분)
nltk.download('punkt')
words = word_tokenize(text)
print(words)
```

```python
from nltk.tokenize import RegexpTokenizer     # 방법 (3) : nltk.tokenize 패키지의 RegexpTokenizer 함수 사용 (=> regular expression을 이용하여 특정 부분 제거 가능)
retokenize = RegexpTokenizer("[\w]+")
words = retokenize.tokenize(text)
print(words)  
```

### Normalization  

* = cleaning. string의 특정 부분을 분리해내거나 추출해내는 것.

**(1) Stemming** 어간 추출 <u>대충의 패턴 규칙으로 어미를 잘라내는 것</u> <b>(사전에 없는 어간 나올 수 있음)</b>


**(2) Lemmatization** 표제어(기본 사전형) 추출. <b> -> 표제어 기반 원형 추출 </b>


```python
# (1) Stemming - stemmer.stem(w) => 사전에 없는 어미가 나오는 경우가 있음 (기계적으로 끊음)

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
[stemmer.stem(w) for w in words]
```

```python
# (1) - 2 : 또 다른 형태의 stemmer => 마찬가지로 표제어가 아닌 어미 추출 하는 경우 있음

from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()
[stemmer.stem(w) for w in words] 
```

```python
# (2) lemmatization : 표제어 기반 원형 어미 추출 (잘라내는 것뿐만 아니라 원형으로 복귀시킴) => 더 정교한 tokenizer

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')                      #wordnet을 기반으로 원형만 추출
lemmatizer = WordNetLemmatizer()
[lemmatizer.lemmatize(w) for w in words]
```

### Stopword

* 불용어

= 단어들 중에 대명사, 조동사, 관사 등 한 text에서 너무 빈번히 나와 중요하지 않은 단어들은 분석에 쓸모 없기 때문에 따로 모아 놓는 기능

```python
#

from nltk.corpus import stopwords  
nltk.download('stopwords')              #regular expression에 정의되어 있는 stopwords를 받아옴
print(words)
words = [w for w in words if not w in stopwords.words('english')] # 단어 하나가 stopwords에 들어 있지 않으면 취해라 = stopwords 단어들을 빼라
print(words)
```

### Collocation, Concordance

* Collocation = 연어

 = ex. 숙어 (take & care / 문장에서 Mike Smith가 빈번히 반복되면 Mike & Smith가 연어)

* Concordance = 특정 단어에 대해서 그 용례가 어떻게 되는지 copus를 통해 보여줌.

 = ex) 'would'가 copus에서 어떻게 사용되는지 용례를 찾아줘.


```python
nltk.download('gutenberg')                              #nltk 패키지 안에 들어 있는 gutenberg 프로젝트의 austen-emma.txt를 가져옴 => type은 string
text = nltk.corpus.gutenberg.raw('austen-emma.txt')
words = retokenize.tokenize(text)                       #단어별로 tokenize -> regular expression based tokenize가 제일 성능이 좋음
```

```python
# Collocation : 연어 (연달아서 나오는 단어들)

nltk.Text(words).collocations()  # default: (num=20, window_size=2) -> 2개짜리 연어(collocation) 총 20개 추출
```

```python
#Concordance: 특정 단어의 용례 찾아줌.

nltk.Text(words).concordance('Emma', 79, 10)   #"Emma"라는 단어가 등장하는 앞 뒤로 79개의 text를, 총 10개 보여줘라
```

```python
#dispersion_plot : corpus에서 특정 text/word가 등장하는 부분을 plot 
#x축은 corpus 전체 

#단어들의 분포를 알 수 있음
#=> 소설 text에서는 누가 주인공이고 조연인지 파악 가능
nltk.Text(words).dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
```

```python
# Distributional similarity: 분포상에서 유사한 특성이 무엇인지
# find other words which appear in the same contexts as the specified word; 
# list most similar words first.

nltk.Text(words).similar("Emma")    #Emma와 비슷하게 양쪽 옆으로 등장하는 단어들을 출력

#Emma는 사람이므로 대명사, 명사들이 추출됨
```

```python
# Find contexts where the specified words appear; list most frequent common contexts first. = 공통으로 나타나는 양쪽 단어들을 찾아라
nltk.Text(words).common_contexts(["Emma", "she"])

#다음과 같은 text에 "Emma"와 "she"과 같이 나타날 가능성이 높다고 해석 가능
```

### Regular expression

```python
import re
```

```python
re.search('a', 'abcdefa')  #re.search(찾을문자열, 검색할 string)

# search() : 왼쪽부터 시작하여 가장 먼저 일치하는 최장문자열을 검색하여 출력 (제일 먼저 일치하는 문자열 하나만 출력) / 출력결과 re.Match object
```

```python
re.findall('a', 'abcdefa')  #re.findall(찾을문자열, 검색할 string)

# findall() : 일치하는 모든 문자열들을 전부 출력 / 출력 결과 리스트
# -> search함수와 다른 점은 일치하는 문자열을 모두 출력한다는 점
```

```python
re.sub('a', 'b', 'abcdefa')   #re.sub(찾을문자열, 바꿀 문자열, 검색할 string)
 # sub() : 찾아 바꾸기 기능 (일치하는것 전부 바꿈). / 출력 결과 문자열
```

```python
#Regular expression 규칙들

'''       Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures

.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]    Matches anything but a set of characters
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	      Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	  Parentheses that indicate the scope of the operators
(...)     Matches whatever regular expression is inside the parentheses
\d
Matches any decimal digit; this is equivalent to the class [0-9].
\D
Matches any non-digit character; this is equivalent to the class [^0-9].
\s
Matches any whitespace character; this is equivalent to the class [ \t\n\r\f\v].
\S
Matches any non-whitespace character; this is equivalent to the class [^ \t\n\r\f\v].
\w
Matches any alphanumeric character; this is equivalent to the class [a-zA-Z0-9_].
\W
Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].

'''
```

```python
engdict = nltk.corpus.words.words('en')            #engdict에 사전 불러와 할당함

# result = [w for w in engdict if re.search('ed$', w)]      # engdict에 있는 단어들 w에 대해(for) 'ed'로 끝나는 단어들이 있으면(if) 찾아라

# result = [w for w in engdict if re.search('^..j..t..$', w)]  # engdict에 있는 단어들 w에 대해 __j__t__인 단어이 있으면 찾아라 (^: 시작 / $: 끝)
#( . -> 모든 word)

# result = [w for w in engdict if re.search('^[ghi][mno][jlk][def]$', w)] #[g or h or i] [m or n or o] [j or l of k] [d or e or f]로 구성된 단어가 있으면 찾아라
# ( [] => 대괄호 안에 들어 있는 단어 중 하나)

result = [w for w in engdict if re.search('^[ah]+$', w)] # [a와 h의 연속으로 구성된 단어]를 찾아라
# ( + => 1번 이상 반복 / cf. * => 0번 이상 반복)

print(result[:10])
```

```python
nltk.download('treebank')            #신문에 있는 단어들을 다운로드 (=사전보다 현실에 가까운 corpus)
wsj = nltk.corpus.treebank.words()

# result = [w for w in wsj if re.search('(ed|ing)$', w)]           # wsj에 있는 단어 중 ed 또는 ing로 끝나는 단어가 있으면 출력해라

#result = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]  #단어 중 0-9의 숫자가 하나이상 /. (\.=점) / 0-9의 숫자 하나이상 단어가 있으면 출력 => 소수 출력
#cf) . => any character / \. => 점


#result = [w for w in wsj if re.search('^[A-Z]+\$$', w)]   #단어 중 A부터 Z 대문자 알파벳 하나 이상/ \$ = $로 끝나는 단어 출력
#cf. $ => 문장 끝 / \$ => $(달러)

#result = [w for w in wsj if re.search('^[0-9]{4}$', w)] #단어 중 0-9 숫자가 4개 나오는 단어 출력 = 숫자 네자리

#result = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)] #단어중 0-9 숫자 한개 이상 / - / [a-z] 3~5개 인 단어 (ex. 10-day)

result = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)] #소문자가 5개 이상 / - / 소문자가 2~3개 / - / 소문자가 6개 이하 단어

result = sorted(set(result))
print(result[:10])
```
