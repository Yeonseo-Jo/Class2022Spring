```python
#dict 형태인 r을 인덱싱하여 data frame 파일 형태로 정리하는 방법.
#인덱싱을 통해 column명 부분이 될 값을 추출해 냄.

items = r['response']['body']['items']

cityName = [item['cityName'] for item in items]  #list comprehension -> itmes에 있는 item 중에 'cityName'을 갖는 값들을 모아라
pm10Value = [item['pm10Value'] for item in items]
pm25Value = [item['pm25Value'] for item in items]
so2Value = [item['so2Value'] for item in items]
coValue = [item['coValue'] for item in items]
o3Value = [item['o3Value'] for item in items]
no2Value = [item['no2Value'] for item in items]

cityName = pd.Series(cityName)
pm10Value = pd.Series(pm10Value)   
pm25Value = pd.Series(pm25Value)
so2Value = pd.Series(so2Value)
coValue = pd.Series(coValue)
o3Value = pd.Series(o3Value)
no2Value = pd.Series(no2Value)

df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})

#pd.Series(column 명에 해당하는 data 값들을 만들어냄)와 pd.DataFrame(dict로 취해줌)이라는 pandas 모듈의 function으로 data frame 형태로 만든다.
```

```python
def test(sidoName):
  URL = ('http://apis.data.go.kr/B552584/ArpltnStatsSvc/'
      'getCtprvnMesureSidoLIst?'
      'sidoName='+sidoName+
      '&searchCondition=hour'
      '&pageNo=1'
      '&numOfRows=100'
      '&returnType=json'
      '&serviceKey=miH%2BZXg85lQy4%2FkmhffvygXDIFiTwisriSLxtyECSzw6hxTVK7yI8QKIOc7mP7LEqCnhBGuj6iW1gZW65dum7A%3D%3D')
  #->원하는 시도이름이 들어오면 그와 관련된 API URL을 가져오도록 연결함.
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  cityName = [item['cityName'] for item in items]       
  pm10Value = [item['pm10Value'] for item in items]
  pm25Value = [item['pm25Value'] for item in items]
  so2Value = [item['so2Value'] for item in items]
  coValue = [item['coValue'] for item in items]
  o3Value = [item['o3Value'] for item in items]
  no2Value = [item['no2Value'] for item in items]

  cityName = pd.Series(cityName)
  pm10Value = pd.Series(pm10Value)   
  pm25Value = pd.Series(pm25Value)
  so2Value = pd.Series(so2Value)
  coValue = pd.Series(coValue)
  o3Value = pd.Series(o3Value)
  no2Value = pd.Series(no2Value)

  df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})
  return df

#gradio는 input-output의 함수 값을 논항 인자로 입력 해야 하므로 def 함수로 입출력의 값을 갖는 함수를 만들어 API를 gradio에 연결한다.
#'sidonname'에 해당하는 input이 들어오면, 다음 함수의 과정을 거쳐 df를 return해주는 사용자 정의 함수, 입출력의 함수를 만든다. (graido에 입출력의 형태로 넣어주기 위해!)
```

```python
#!pip install gradio
#import gradio as gr
#=> gradio는 colab에 기본적으로 install 된 모듈이 아니기 때문에 !pip install로 install을 해 준 다음 import를 해야 함.


iface = gr.Interface(test, gr.inputs.Dropdown(['서울', '부산', '대구', '인천', '광주', '대전', '울산', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주', '세종']), outputs=['dataframe'])
iface.launch(debug=True)
#gradio.inputs.Dropdown -> dropdown list의 형식으로 입력 값을 넣음 / output은 그냥 dataframe 형식.


#만들었던 dataframe을 gradio에 심어 만든 data frame을 웹앱의 형식 demo로 만들 수 있다.
#gradio는 일종의 input-output function의 형태로 넣어야 함. -> def로 사용자정의함수를 만들어 input-output 값을 지정해준다.
```
