```python
!pip install gradio

import gradio as gr

import requests 
#API를 가져올 수 있는 파이썬 라이브러리: requests

import pandas as pd
```

```python
#API 불러오는 코드: requests.get(서비스키URL)
#인터넷 URL을 주면, API 서버에서 응답을 받아오는 방식
#가져올 API의 매뉴얼에 다 나와있음
```

```python
URL = ('http://apis.data.go.kr/B552584/ArpltnStatsSvc/'    #=>서비스 URL
      'getCtprvnMesureSidoLIst?'   #=>구체적인 세부 정보의 서비스 이름
      'sidoName=서울'              #=>세부 정보를 가져오기 위한 요청 항목 메세지들
      '&searchCondition=hour'
      '&pageNo=1'
      '&numOfRows=100'
      '&returnType=json'           #=>return 받을 파일의 형태 선택(ex.xml, json)
      '&serviceKey=miH%2BZXg85lQy4%2FkmhffvygXDIFiTwisriSLxtyECSzw6hxTVK7yI8QKIOc7mP7LEqCnhBGuj6iW1gZW65dum7A%3D%3D')
response = requests.get(URL)
```

```python
#dict 구조 => z{key:values}
#dict 인덱싱 => z[key] => value값 인덱싱 됨

#import pandas as pd
items = r['response']['body']['items']      #->value를 불러올 key 값들

#print(items)
#print(items[0])

cityName = [item['cityName'] for item in items]       
#listcomprehension 식: for 루프를 돌려 items에 있는 25개의 item들 중 'cityName'이 있는 것들의 value값들을 CityName 변수에 할당해라
pm10Value = [item['pm10Value'] for item in items]
pm25Value = [item['pm25Value'] for item in items]
so2Value = [item['so2Value'] for item in items]
coValue = [item['coValue'] for item in items]
o3Value = [item['o3Value'] for item in items]
no2Value = [item['no2Value'] for item in items]
#위에 city name, pm10Value 등은 data frame 파일에서 column명(=dataframe의 index)이 될 것들

cityName = pd.Series(cityName)
pm10Value = pd.Series(pm10Value)   
pm25Value = pd.Series(pm25Value)
so2Value = pd.Series(so2Value)
coValue = pd.Series(coValue)
o3Value = pd.Series(o3Value)
no2Value = pd.Series(no2Value)
#pd.Series => 각 column명에 맞는 column들을 만듦.

df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})
#pandas 모듈의 dataFrame 메소드를 이용하여 df 변수명에 DataFrame 파일(=엑셀 파일과 같은 형태)을 생성하고 지정함.
#형태 = {column명: column명에 맞는 내용}
```

```python

#import gradio as gr
#=>gradio에 심어서 웹사이트에 정보를 넘김
#gradio: Gradio는 머신러닝 모델을 웹 앱 형태로 빌드하고 배포할 수 있도록 돕는 파이썬 라이브러리이다

def test(sidoName):
  URL = ('http://apis.data.go.kr/B552584/ArpltnStatsSvc/'
      'getCtprvnMesureSidoLIst?'
      'sidoName='+sidoName+
      '&searchCondition=hour'
      '&pageNo=1'
      '&numOfRows=100'
      '&returnType=json'
      '&serviceKey=miH%2BZXg85lQy4%2FkmhffvygXDIFiTwisriSLxtyECSzw6hxTVK7yI8QKIOc7mP7LEqCnhBGuj6iW1gZW65dum7A%3D%3D')
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']

  cityName = [item['cityName'] for item in items]
  pm10Value = [item['pm10Value'] for item in items]
  pm25Value = [item['pm25Value'] for item in items]
  so2Value = [item['so2Value'] for item in items]
  coValue = [item['coValue'] for item in items]
  o3Value = [item['o3Value'] for item in items]
  no2Value = [item['no2Value'] for item in items]

  cityName = pd.Series(cityName)
  pm10Value = pd.Series(pm10Value)   
  pm25Value = pd.Series(pm25Value)
  so2Value = pd.Series(so2Value)
  coValue = pd.Series(coValue)
  o3Value = pd.Series(o3Value)
  no2Value = pd.Series(no2Value)

  df = pd.DataFrame({ 'cityName': cityName, 'pm10Value': pm10Value, 'pm25Value': pm25Value, 'so2Value': so2Value, 'coValue': coValue, 'o3Value': o3Value, 'no2Value': no2Value})
  return df
```

```python
iface = gr.Interface(test, gr.inputs.Dropdown(['서울', '부산', '대구', '인천', '광주', '대전', '울산', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주', '세종']), outputs=['dataframe'])
iface.launch(debug=True)
```
