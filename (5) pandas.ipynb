### (1) Constructing a DataFrame using dict

#### Data frame을 만드는 두번째 방법 : dict를 통해 직접 값을 입력하여 dataframe 파일을 만드는 방법.

```python
# Some input data - a small sample of the iris dataset

#data라는 변수의 type = dict!
# *dict는 {}로 만들어지는 데이터 타입으로, key 값과 value 값이 콜론으로 연결되어 있음. (key와 value의 개수가 맞아야 함.)

data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],          #->이게 한 column! (*그래서 개수가 맞아야 함.)
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

# => data를 dict를 통해 만들었음. / key 값들이 column명, value 값들이 그 column에 해당하는 데이터 값들.

df = pd.DataFrame(data)
df 

#pandas에 있는 .DataFrame() 함수를 이용해서 data frame을 만듦. / data라는 dict를 입력 값으로 넣어 이를 바탕으로 한 data frame을 만든다.
#csv 파일로 data frame을 만들 수도 있지만, data frame의 형식(=엑셀 파일과 같은 형식)을 가지는 dict를 통해 직접 값들을 넣어주고 data frame을 만드는 방법도 있음.
```

### (3) Constructing a DataFrame using pd.Series
#### Data frame을 만드는 세번째 방법: pd.Series 함수 이용하기

```python
#pd.Series()로 데이터 값들을 먼저 만들어 준 다음, 그 후에 dict로 만들어주는 방법.

sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])
#총 5개의 column, 총 5개의 row

#pd.Series로 dataframe을 만드는 방법 = pd.Series()의 입력 값으로 한 column씩 값을 지정해줌.
# 형식 : 변수명(=column 명) = pd.Series([각각의 row에 들어갈 data 값들])


df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
df

#pd.DataFrame의 입력 값에 들어갈 dict 값을 앞에서 만든 pd.Series를 이용하여 만들어줌.
```

### Changing column names 
#### = data frame에서 column명을 바꾸는 방법

```python
# Change the columns
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df

# .columns 함수: column명을 바꾸는 방법.
# 변수명.columns = 변경할 column명을 포함한 column명의 리스트 (여기서는 대문자였던 SPECIES를 소문자 species로 바꿈)
```

### Accessing columns
####: 특정 column에만 access하는 방법

###### 1) df.loc[row(:으로 전체 범위로 지정), column]
###### 2) df[column명]
###### 3) df.column명

```python
# Access one column 
df.loc[:, 'sepal_length']

# 첫번째 방법 : data frame을 저장한 변수명(=df).loc[row,column] -> 전체 row와 (: 을 사용했기 때문에 모든 행을 불러온다는 것을 뜻함.) 'sepal_length'에 해당하는 column의 정보만 acessing 한다.
```

```python
# Access one column (short-hand)
df['sepal_length']

#두번째 방법: df['column명']
```

```python
# Access one column (shortest-hand)
df.sepal_length

#세번째 방법: df.column명  (-> 얘는 one column만 가능)
```

```python
# Access two columns: 두개의 column 가져오는 방법
df.loc[:, ['sepal_length', 'sepal_width']]

#첫번째 방법과 동일 -> column명에 해당하는 부분을 두 개 입력해주면 됨.
```

```python
# Access two columns (short-hand)

df[['sepal_length', 'sepal_width']]

#두번째 방법과 동일 -> column명에 해당하는 부분을 두 개 입력해주면 됨
```

### Accessing rows : 특정 행에만 acess하는 방법
##### 1) df.loc[row(찾을 행의 number), column (:로 전체 column 지정)]
##### 2) df.loc과 리스트,슬라이싱, 조건식을 이용해서 특정 범위의 행에만 접근할 수도 있다.

```python
# Access one row 
df.loc[3, :]

#-> 세번째 row 한 개만 통으로 가져오는 access 방법
```

```python
# Access two rows
df.loc[[3, 0], :]

#-> df.loc[]와 리스트를 이용해서 3번째 행과 0번째 행 두 개를 가져오는 방법
```

```python
# Access a range of rows
df.loc[2:5, :]

#-> df.loc[]와 인덱싱을 이용해서 2번째 row부터 4번째 row(슬라이싱에서는 -1까지여서 5번째가 아닌 4번째까지임!!) 까지의 정보를 불러오는 방법
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]

#->sepal_length의 값이 5보다 큰 경우에만 그 행을 가져와라 => 조건에 true가 되는 0,1,3번째 row에만 access함.
```

### Accessing a cell

#### -> row, column이 아닌 하나의 cell에만 acessing하는 방법.
###### column or row accessing 방법으로 one column or row만 불러오고 column 인덱싱 이용하기

```python
df.loc[:, 'sepal_length'][4]
#-> 'sepal_length' column에서 4번째에 해당하는 값만 가져와라 = 4.6 (*주의! 인덱싱은 0부터 시작.)
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

### Adding/Removing new columns
##### = 한 column이나 row를 더하거나 빼는 방법

```python
df['new_column'] = 1 #-> column을 accessing하는 방법 (변수명[column명])을 이용해서 새로운 column 만드는 방법
df


#-> new_column을 새로 추가하는 방법
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df

#-> column acessing 방법으로 new_column의 값들을 바꾸는 방법! = 리스트 이용
```

```python
df = df.drop('new_column', axis=1) # axis=1: column  / (cf. axis=0 : row)
df

#column을 제거하는 방법: df.drop()
#axis=1 : column 정보이다 / 즉, column에서 해당 변수명에 해당하는 column을 찾아 제거해라
#axis=0 : row 정보이다 / 즉, row에서 해당 변수명에 해당하는 row를 찾아 제거해라
```

### Adding/Removing new rows
###### : 새로운 row를 더하거나 빼는 방법.

```python
#새로운 행을 만드는 방법

df.loc[5, :] = ''
df

#df.loc[row, column]
# -> row에는 새로운 row가 추가될 행의 값, column은 전체로 지정하고, 그 값을 = 을 통해 지정해 줌. / 여기서는 '', blank로 지정해주었기 때문에 값이 blank인 5번째 행이 새로 추가됨.
```

```python
#특정 행을 지우는 방법

df = df.drop(5, axis=0) # axis=0: row
df

#-> df.drop을 이용./ column과 방법은 같지만 row의 경우에는 axis=0을 해줘야 함.
# -> df.drop을 이용해서 행을 제거한 값을 다시 df 변수명에 저장해 줌. (다시 저장해 줘야 변경한 값이 변수에 저장됨)
```
