```python
def test(RGN_SUB):
  URL =("http://apis.data.go.kr/B553530/TRANSPORTATION/ELECTRIC_CHARGING?servicekey=Mf9iEGO3qN4SJfdvGJ3LeoFfkx6SIW86RRfJNDsr44xsB2yw6eWf9nQ/i2JbxjarI9s2e3YqFqZlzRPkeZNBcQ==&pageNo=1&numOfRows=100&apiType=json&q1=강원도&q2="+RGN_SUB)
  
  response = requests.get(URL)
  r=response.json()

  items = r['opentable']['field']
  RGN_MST = [item['RGN_MST'] for item in items]
  RGN_SUB = [item['RGN_SUB'] for item in items]
  ITEM_6 = [item['ITEM_6'] for item in items]
  DATA_REG_DT = [item['DATA_REG_DT'] for item in items]
  
  RGN_MST = pd.Series(RGN_MST)
  RGN_SUB = pd.Series(RGN_SUB)
  ITME_6 = pd.Series(ITEM_6)
  DATA_REG_DT = pd.Series(DATA_REG_DT)
  
  df=pd.DataFrame({'RGN_MST' : RGN_MST,'RGN_SUB': RGN_SUB, 'ITEM_6' : ITEM_6, 'DATA_REG_DT' : DATA_REG_DT})
  df.columns=['광역지자체(강원도)','시군구','사용량','자료등록일시' ]
  return df

  
```
